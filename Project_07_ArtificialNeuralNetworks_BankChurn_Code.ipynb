```python
# Defining a Function to plot Discrete Featues vs Target Feature:

def stacked_barplot(data, predictor, target):
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True) \
                      .sort_values(by=sorter, ascending=False)

    print('\033[0m' , '\033[92m' , 'Stacked Bar Plot: ' , 
      '\033[1m' , '\033[91m' , predictor , 
      '\033[0m' , '\033[92m' , ' vs ' ,
      '\033[1m' , '\033[91m' , target , 
      '\033[0m' , '\033[92m' , sep='')    

    print(tab1)
    tab = pd.crosstab(data[predictor], data[target], normalize='index') \
                .sort_values(by=sorter, ascending=False)
    tab.plot(kind='bar', stacked=True, color = ('Lime','yellow'), figsize=(count + 5, 4))
    plt.legend(loc='lower left', frameon=False)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.xlabel('')
    plt.show() 
```

```python
# Defining a Function to plot Continuous Featues vs Target Feature:

def continuous_plot(x , y):
    print('\033[0m' , '\033[92m' , 'Box Plot: ' , 
          '\033[1m' , '\033[91m' , x , 
          '\033[0m' , '\033[92m' , ' vs ' ,
          '\033[1m' , '\033[91m' , y , 
          '\033[0m' , '\033[92m' , sep='') 
        
    values = df[y].unique().tolist()
    
    for value in values:
        mean = round(np.mean(df[df[y]==value][x]) , 2)
        print('\033[0m' , '\033[92m' , 'Class ' , 
              '\033[1m' , '\033[91m' , value , 
              '\033[0m' , '\033[92m' , ' Mean = ' ,
              '\033[0m' , '\033[92m' , f"{mean:,}" , 
              '\033[0m' , '\033[92m' , sep='')

    sns.catplot(data=df , x=y , y=x , kind='box' , ci=False , height=4 , \
                showfliers=False , color = 'Lime')
    plt.show();
```

<left> <font size=5 style='font-family:Courier New' color='Green'> **Good to know!**
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **1. Definitions:**
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **1.1:** Sensitivity (Recall/True Positive Rate): Indicates how well the Model can capture the positive Classes and avoid False Negatives.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **1.2:** False Positive Rate: Indicates the Model tendency to make False Positive Errors (Classifying Negative Classes as Positive) .
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **1.3:** Specificity: Indicates how well the Model can avoid classifying Negative Classes as Positive.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **1.4:** Geometric Mean (G-Mean): Metric for Imbalanced Classification. If optimized, will seek a balance between the Sensitivity and the Specificity.

<br> <left> <font size=3 style='font-family:Courier New' color='black'> **2. Equations:**
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **2.1:** True Positive Rate (TPR) = True Positive / (True Positive + False Negative)
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **2.2:** False Positive Rate (FPR) = False Positive / (False Positive + True Negative)
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **2.3:** Specificity = 1 - False Positive Rate = True Negative / (False Positive + True Negative)
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **2.4:** G-Mean = sqrt(Sensitivity * Specificity) = sqrt((TP / (TP + FN)) * (TN / (TN + FP))).

<br> <left> <font size=3 style='font-family:Courier New' color='black'> In ROC Curve, the Threshold with the largest G-Mean value is chosen.

```python
# Defining a Function to Plot the Confusion Matrix:

def plot_confusion_matrix(cf , group_names=None , categories='auto' , xyplotlabels=True , sum_stats=True ,
                          count=True , percent=True , cbar=True , xyticks=True , figsize=None , cmap='Blues' , title=None):

    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ['{}\n'.format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ['{0:0.0f}\n'.format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ['{0:.2%}'.format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f'{v1}{v2}{v3}'.strip() for v1, v2, v3 in zip(group_labels , group_counts , group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0] , cf.shape[1])

    if sum_stats:
        accuracy  = np.trace(cf) / float(np.sum(cf))

        if len(cf)==2:
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)

        else:
            stats_text = ''
    else:
        stats_text = ''

    if figsize==None:
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        categories=False

    plt.figure(figsize=figsize)
    sns.heatmap(cf , annot=box_labels , fmt='' , cmap=cmap , cbar=cbar , xticklabels=categories , yticklabels=categories)
    
    if xyplotlabels:
        plt.ylabel('Actual Class')
        plt.xlabel('Predicted Class')
              
    if len(cf)==2:
        print('\033[0m' , '\033[92m' , 'Accuracy = ' ,
              '\033[1m' , '\033[91m' , '{0:.3f}'.format(accuracy) , 
              '\033[0m' , '\033[92m' , '\nPrecision = ' ,
              '\033[1m' , '\033[91m' , '{0:.3f}'.format(precision) ,
              '\033[0m' , '\033[92m' , '\nRecall = ' ,
              '\033[1m' , '\033[91m' , '{0:.3f}'.format(recall) ,              
              '\033[0m' , '\033[92m' , '\nF1 Score = ' ,
              '\033[1m' , '\033[91m' , '{0:.3f}'.format(f1_score) , sep='')              
    else:
        print('\033[0m' , '\033[92m' , 'Accuracy = ' ,
              '\033[1m' , '\033[91m' , '{0:.3f}'.format(accuracy) , sep='')
```

```python
# Defining a Function to Plot the ROC-AUC Curve:

def plot_roc_auc_curve(model , predictiors , validators):
    yhat_2 = model.predict(predictiors)

    yhat_2 = yhat_2[: , 0]
    fpr , tpr , thresholds1 = roc_curve(validators , yhat_2)
    gmeans1 = np.sqrt(tpr * (1-fpr))

    ix = np.argmax(gmeans1)
    
    global threshold_roc
    threshold_roc = thresholds1[ix]
    
    print('\033[0m' , '\033[92m' , 'Optimum Threshold = ' ,
          '\033[1m' , '\033[91m' , '{0:.6f}'.format(threshold_roc) ,
          '\033[0m' , '\033[92m' , ' at G-Mean = ' ,
          '\033[1m' , '\033[91m' , '{0:.3f}'.format(gmeans1[ix]) , sep='')

    plt.plot([0,1] , [0,1] , linestyle='--' , label='No Skill' , color='orange')
    plt.plot(fpr , tpr , marker='.' , color='orangered')
    plt.scatter(fpr[ix] , tpr[ix] , marker='o' , color='purple' , label='Best' , s=100)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
```

```python
# Clearing the Previous Sessions:

backend.clear_session()
np.random.seed(42)
import random
random.seed(42)
tf.random.set_seed(42)
```

```python
# Defining the Model's Function:

def model_x(drop_out , learning_rate):  
    np.random.seed(1337)
    
    model = Sequential()
    
    model.add(Dense(units=128 , activation = 'relu' , input_shape=(X_train.shape[1],)))
    
    model.add(Dense(units=64 , activation='relu'))
    model.add(Dropout(drop_out))
    
    model.add(Dense(units=32 , activation='relu'))
    
    model.add(Dense(units=1 , activation = 'sigmoid')) 

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss= 'binary_crossentropy' , optimizer=optimizer , metrics=['accuracy'])  

    return model
```

```python
# Defining the Grid Search Scope:

keras_estimator = KerasClassifier(build_fn=model_x, verbose=1)
param_grid = {'drop_out' : [0.1 , 0.2 , 0.3] , 
              'learning_rate' : [0.001 , 0.01 , 0.1] ,
              'batch_size' : [32 , 64 , 128]}

kfold_splits = 5
grid = GridSearchCV(estimator=keras_estimator , verbose=1 , cv=kfold_splits , param_grid=param_grid , n_jobs=-1)
```
