<br> <br> <left> <font size=3 style='font-family:Courier New' color='Black'>**Revolving Balance:** If we don't pay the balance of the revolving credit account in full every month, the unpaid portion carries over to the next month. That's called a revolving balance.

<br> <left> <font size=3 style='font-family:Courier New' color='Black'>**Average utilization Ratio:** Represents how much of the available credit the customer spent. This is useful for calculating credit scores.
<br> <br> <left> <font size=3 style='font-family:Courier New' color='Blue'>**(Avg_Open_To_Buy/Credit_Limit) + Avg_Utilization_Ratio = 1**

<left> <font size=6 style='font-family:Courier New' color='Orange' > **Case Understanding:**
<br> <left> <font size=3 style='font-family:Courier New' color='black'> In order to help Thera Bank maintain customer behavior in terms of frequency usage of its credit cards, we need to understand the different customer attributes, draw an initial insights of the possible effect of each, do the necessary analysis, and finally building a model that might help in predicting the Attrited customer and accordingly, do the necessary preventive actions to avoid such renouncation as much as possible.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> Initially, we can observe that some attributes might lead into an increase in customer obligations; Such increase in obligations will result in an increase in spending using the Credit Cards:
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Dependent_count**: More dependents leads to an increase in obligations.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Marital_Status**: Usually, Married have more obligations than Singles.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Income_Category**: Logically, higher income results in higher obligations.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> On the other hand, some attributes could be linked to customer behaviour; By understanding the behaviour, you might expect if the customer is prone to renounce the credit card or not: 
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Total_Trans_Amt**: The more the Transactions Amount, the more the spending, the more is the customer in real need for credit card. Furthermore, some correlation is expected between this attribute and Total_Trans_Ct.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Months_Inactive_12_mon**: the more the customer is inactive, the more the probability of Attrition.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Total_Revolving_Bal**: The more the Revolving Balance, the less is the amount left for usage, the less is the spending. On the other hand, the less the Revolving Balance, the more the customer is committed to repay.
<br> <left> <font size=3 style='font-family:Courier New' color='black'> **Total_Amt_Chng_Q4_Q1**: Indicates if there is an increase in customer spending or not. The less means that the customer tends to renounce the credit card due to less usage.

```python
# function to plot stacked bar chart

def stacked_barplot(data, predictor, target):
    """
    Print the category counts and plot a stacked bar chart

    data: dataframe
    predictor: independent variable
    target: target variable
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    tab.plot(kind="bar", stacked=True, figsize=(count + 1, 5))
    plt.legend(
        loc="lower left", frameon=False,
    )
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()
```

```python
# Defining a Function to plot the Continuous Featues:

def continuous_plot(x , y):
    print('\033[1m' , '\033[92m' , 'Box Plot (' , '\033[91m'  , x , ' vs ', y , '\033[92m' , '):' , 
          '\033[0m' , '\033[92m' , sep='')
    values = df[x].unique().tolist()
    for value in values:
        mean = round(np.mean(df[df[x]==value][y]) , 2) 
        print('\033[1m' , '\033[94m' , 'Mean (' , value  , ') = ' , f'{mean:,}' , sep = '')
    sns.catplot(data=df , x=x , y=y , kind='box' , ci=False , height=4 , \
                showfliers=False , color = 'Lime')
    plt.tick_params(axis='x', rotation=90)
    plt.show();
```

```python
# Defining a Function to Plot the Confusion Matrix:

def confusion_matrix_plot(model , predictors , target):
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target , y_pred)
    labels = np.asarray([['{0:0.0f}'.format(item) + '\n{0:.2%}'.format(item / cm.flatten().sum())]
             for item in cm.flatten()]).reshape(2, 2)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm , annot=labels , fmt="" , cmap='viridis')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
```

```python
# Defining a Function to calculate the Performance Metrics:
def performance_metrics(model , predictors , target):
    pred = model.predict(predictors)

    acc = accuracy_score(target , pred)
    recall = recall_score(target , pred)
    precision = precision_score(target , pred)
    f1 = f1_score(target , pred)

    df_perf = pd.DataFrame({'Accuracy' : acc , 
                            'Recall' : recall ,
                            'Precision' : precision ,
                            'F1 Score' : f1} , index=[0])
    return df_perf
```

```python
# Dataset Preprocessing:

df_pipe.drop(['CLIENTNUM'] , axis = 1 , inplace=True)

df_pipe['Income_Category'].replace('abc', np.nan, inplace=True)

df_pipe["Attrition_Flag"].replace("Existing Customer", 0, inplace=True)
df_pipe["Attrition_Flag"].replace("Attrited Customer", 1, inplace=True)
```

```python
# Creating a Transformer for Numerical Features:

numerical_features = ['Customer_Age' , 'Months_on_book' , 'Total_Relationship_Count' , 'Months_Inactive_12_mon' ,
                      'Contacts_Count_12_mon' , 'Credit_Limit' , 'Total_Revolving_Bal' , 'Avg_Open_To_Buy' , 
                      'Total_Amt_Chng_Q4_Q1' , 'Total_Trans_Amt' , 'Total_Trans_Ct' , 'Total_Ct_Chng_Q4_Q1' , 
                      'Avg_Utilization_Ratio']

numerical_features_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median'))])
```

```python
# Creating a Transformer for Categorical Features:

categorical_features = ['Gender','Dependent_count' , 'Education_Level' , 'Marital_Status' , 'Income_Category' , 
                        'Card_Category']

categorical_features_transformer = Pipeline(steps=[('imputer' , SimpleImputer(strategy='most_frequent')) ,
                                          ('encoder', OneHotEncoder(handle_unknown='ignore'))])
```

```python
# Creating a Column Transformer:

col_transformer = ColumnTransformer(transformers=[('num_trans' , numerical_features_transformer , numerical_features) ,
                                                  ('cat_trans', categorical_features_transformer, categorical_features)] ,
                                                  remainder="drop")
```

```python
# Splitting the Dataset into Independent and Dependent Subsets:

X = df_pipe.drop('Attrition_Flag' , axis=1)
y = df_pipe.pop('Attrition_Flag')
```

```python
'''
Splitting the Dataset into Training and Testing Subsets:
Validation Subset is no longer needed, Training Subset has been already validated and Hyperparameters Tuning is already done!
We're ready to deploy the Pipeline on Testing Subset and/or Unseen Datasets.
'''

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=1, stratify=y)

print('\033[0m' , '\033[92m' , '1. Training Subset Shape: ', '\033[1m' , '\033[92m' , X_train.shape , sep='')
print('\033[0m' , '\033[92m' , '3. Testing Subset Shape: ', '\033[1m' , '\033[92m' , X_test.shape , sep='')
```

```python
# Under-Sampling the Dataset using RandomUnderSampler:

rus = RandomUnderSampler(random_state=1, sampling_strategy=1)
X_train_un, y_train_un = rus.fit_resample(X_train, y_train)
```

```python
# Displaying the contents of both X and y before & after Under-Sampling:

print('\033[1m' , '\033[92m' , '1. Training Subset (Before Under-Sampling):' , '\033[0m' , sep='')
print('\033[92m' , '  - No. of False Cases (0): {0} ({1:0.2f}%)'. format(len(y_train[y_train[:]==0]) , 
     (len(y_train[y_train[:]==0])/len(y_train)) * 100))
print('\033[92m' , '  - No. of True Cases (1): {0} ({1:0.2f}%)'. format(len(y_train[y_train[:]==1]) , 
     (len(y_train[y_train[:]==1])/len(y_train)) * 100) , '\n')

print('\033[1m' , '\033[92m' , '2. Training Subset (After Under-Sampling):' , '\033[0m' , sep='')
print('\033[92m' , '  - No. of False Cases (0): {0} ({1:0.2f}%)'. format(len(y_train_un[y_train_un[:]==0]) , 
     (len(y_train_un[y_train_un[:]==0])/len(y_train_un)) * 100))
print('\033[92m' , '  - No. of True Cases (1): {0} ({1:0.2f}%)'. format(len(y_train_un[y_train_un[:]==1]) , 
     (len(y_train_un[y_train_un[:]==1])/len(y_train_un)) * 100) , '\n')

print('\033[1m' , '\033[92m' , '3. Testing Subset:' , '\033[0m' , sep='')
print('\033[92m' , '  - No. of False Cases (0): {0} ({1:0.2f}%)'. format(len(y_test[y_test[:]==0]) , 
     (len(y_test[y_test[:]==0])/len(y_test)) * 100))
print('\033[92m' , '  - No. of True Cases (1): {0} ({1:0.2f}%)'. format(len(y_test[y_test[:]==1]) , 
     (len(y_test[y_test[:]==1])/len(y_test)) * 100) , '\n')

print('\033[1m' , '\033[92m' , '4. Shapes:' , '\033[0m' , sep='')
print('\033[0m' , '\033[92m' , '  - X_train Shape (Before Uner-Sampling): ', '\033[0m' , '\033[92m' , X_train.shape , sep='')
print('\033[0m' , '\033[92m' , '  - y_train Shape (Before Uner-Sampling): ', '\033[0m' , '\033[92m' , y_train.shape , sep='')
print()
print('\033[0m' , '\033[92m' , '  - X_train Shape (After Under-Sampling): ', '\033[0m' , '\033[92m' , X_train_un.shape , sep='')
print('\033[0m' , '\033[92m' , '  - y_train Shape (After Uner-Sampling): ', '\033[0m' , '\033[92m' , y_train_un.shape , sep='')
print()
print('\033[0m' , '\033[92m' , '  - X_test Shape: ', '\033[0m' , '\033[92m' , X_test.shape , sep='')
print('\033[0m' , '\033[92m' , '  - y_test Shape: ', '\033[0m' , '\033[92m' , y_test.shape , sep='')
```

```python
# Building the Pipeline:

pipe = Pipeline(steps=[('imputation_encoding' , col_transformer) ,
                       ('tuned_gboosting' , GradientBoostingClassifier(init=RandomForestClassifier(random_state=1) , 
                                                                                                  subsample=0.8 , n_estimators=450 , max_features=1.0 , 
                                                                                                  learning_rate=0.1 , random_state=1))])
pipe.fit(X_train_un, y_train_un)
```
